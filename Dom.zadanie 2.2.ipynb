```python
import pandas as pd
```

```python
df = pd.read_csv( "data/Minjust2018.csv", 
                 header = None, 
                 usecols=[4,16,19,27,28,33])

df.rename(columns={4: "name",
                   16: "street",
                   19: "tel",
                   27: "director",
                  28: "activity",
                  33: "founders"},
         inplace=True)
df.fillna('', inplace = True)
```

```python
df
```

```python
#1 Найти все компании, которые находятся на вашей улице

df[df['street'].str.contains('Саткыналиева')]
```

```python
#2 Найти все компании, которые принадлежат вашим тезкам

df[df['founders'].str.contains('Акыл')]
```

```python
#3 Найти все компании, которые находятся на вашей улице и принадлежат вашим тезкам

df[df['street'].str.contains('Саткыналиева') & df['founders'].str.contains('Акыл')]
```

```python
#4 Найти две компании, принадлежащие вашим друзьям, знакомым, родителям

df[df['founders'].str.contains('Тимурбек уулу Элдияр') |
df['founders'].str.contains('Семетеев Тилек')]
```

```python
#5 Найти все компании, которые занимаются строительством и в своем телефоне имеют цифры 432 подряд

df[df['activity'].str.contains('строительство') &
df['tel'].str.contains('432')]
```

```python
#6.1 Все задания, но выбрать только название компании, адрес, директора.
#(изначально берем все колонки, которые были сегодня + адрес + телефон)

df.loc[df['street'].str.contains('Саткыналиева'), ['name','street','director']]
```

```python
#6.2 Все задания, но выбрать только название компании, адрес, директора.
#(изначально берем все колонки, которые были сегодня + адрес + телефон)

df.loc[df['founders'].str.contains('Акыл'), ['name','street','director']]
```

```python
#6.3 Все задания, но выбрать только название компании, адрес, директора.
#(изначально берем все колонки, которые были сегодня + адрес + телефон)

df.loc [df['street'].str.contains('Саткыналиева') & df['founders'].str.contains('Акыл'), ['name','street','director']]
```
